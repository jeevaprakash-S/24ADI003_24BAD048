```python
sales_per_product = df.groupby("Description")["Quantity"].sum().head(10)
```

```python
plt.figure()
sales_per_product.plot(kind="bar")
plt.title("Top 10 Products by Sales Quantity")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
plt.figure()
sales_per_product.plot(kind="line")
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.show()
```
